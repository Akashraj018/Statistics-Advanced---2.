1. What is hypothesis testing in statistics?

Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on a sample of data. It helps determine whether there is enough evidence in the sample data to support a specific claim (called a hypothesis) about the population.
Hypothesis testing allows you to use sample data to make objective decisions about claims or theories. It's widely used in science, business, medicine, and many other fields to validate assumptions and guide action based on evidence.

Main Concepts of hypothesis testing
* Null Hypothesis
* Alternative Hypothesis
* Significance Level
* Test Statistic
* p-value

3.Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

The significance level, denoted by alpha, is the threshold for deciding whether to reject the null hypothesis (Ho).
It represents the maximum probability you are willing to accept of making a Type I error — which means rejecting the null hypothesis when it is actually true.
After performing a test (like a t-test or z-test), you calculate a p-value — the probability of observing your sample result (or more extreme) if Ho were true.

The lower your alpha, the stricter your test — you're requiring stronger evidence to reject Ho.
But a very low α increases the chance of failing to detect a real effect (Type II error).

```python
# Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import random
import matplotlib.pyplot as plt

# Parameters
n = 10       # Number of trials
p = 0.5      # Probability of success
size = 10000  # Number of experiments

# Generate binomial data manually
binomial_data = []
for _ in range(size):
    success_count = 0
    for _ in range(n):
        if random.random() < p:
            success_count += 1
    binomial_data.append(success_count)

# Plot histogram
plt.hist(binomial_data, bins=range(n+2), edgecolor='black', align='left', density=True)
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.xticks(range(n+1))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Implement hypothesis testing using Z-statistics for a sample dataset in
#Python. Show the Python code and interpret the results.
#sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
#50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
#50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
# 50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy.stats import norm

# Sample data
sample_data = [
    49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
    50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
    50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
    50.3, 50.4, 50.0, 49.7, 50.5, 49.9
]

# Hypothesis parameters
mu_0 = 50         # Population mean
sigma = 0.5       # Population standard deviation
alpha = 0.05      # Significance level

# Sample statistics
sample_mean = np.mean(sample_data)
n = len(sample_data)

# Z-statistic calculation
z = (sample_mean - mu_0) / (sigma / np.sqrt(n))

# Two-tailed p-value
p_value = 2 * (1 - norm.cdf(abs(z)))

# Print results
print(f"Sample mean: {sample_mean:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Hypothesis test decision
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from 50.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from 50.")
```

```python
# Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib.
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Step 1: Simulate data from a normal distribution
np.random.seed(42)  # For reproducibility
mu = 100            # True population mean
sigma = 15          # Population standard deviation
n = 50              # Sample size

data = np.random.normal(loc=mu, scale=sigma, size=n)

# Step 2: Calculate sample mean and standard error
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # Use ddof=1 for sample std
standard_error = sample_std / np.sqrt(n)

# Step 3: Calculate 95% confidence interval
confidence = 0.95
z_critical = stats.norm.ppf(1 - (1 - confidence) / 2)  # z for 95% CI ~1.96

margin_of_error = z_critical * standard_error
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Step 4: Display results
print(f"Sample Mean = {sample_mean:.2f}")
print(f"95% Confidence Interval = ({ci_lower:.2f}, {ci_upper:.2f})")

# Step 5: Plot the data
plt.figure(figsize=(10, 6))
plt.hist(data, bins=10, edgecolor='black', alpha=0.7, color='skyblue')
plt.axvline(sample_mean, color='red', linestyle='--', label=f'Mean = {sample_mean:.2f}')
plt.axvline(ci_lower, color='green', linestyle='--', label=f'95% CI Lower = {ci_lower:.2f}')
plt.axvline(ci_upper, color='green', linestyle='--', label=f'95% CI Upper = {ci_upper:.2f}')
plt.title("Histogram of Simulated Normal Data with 95% Confidence Interval")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
#Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean

import numpy as np
import matplotlib.pyplot as plt

def calculate_z_scores(data):
    """
    Calculate Z-scores for a dataset.

    Parameters:
        data (list or array): The original dataset.

    Returns:
        z_scores (array): Standardized Z-scores.
    """
    mean = np.mean(data)
    std = np.std(data, ddof=1)  # ddof=1 for sample standard deviation
    z_scores = (data - mean) / std
    return z_scores

# Simulate some data (e.g. normal distribution)
np.random.seed(42)
data = np.random.normal(loc=100, scale=15, size=100)

# Calculate Z-scores
z_scores = calculate_z_scores(data)

# Print first 5 Z-scores
print("First 5 Z-scores:", np.round(z_scores[:5], 2))

# Plot histogram of Z-scores
plt.figure(figsize=(10, 6))
plt.hist(z_scores, bins=10, edgecolor='black', color='lightcoral', alpha=0.7, density=True)
plt.title('Histogram of Z-scores (Standardized Data)')
plt.xlabel('Z-score')
plt.ylabel('Density')
plt.axvline(0, color='blue', linestyle='--', label='Mean (Z=0)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()
plt.show()
```
